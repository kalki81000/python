# Isolation **forest**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/content/archive (1).zip")
```

```python
df.head()
```

```python
plt.scatter(df.iloc[: , 0] , df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
out = IsolationForest(contamination=0.2)
out.fit(df)
pred = out.predict(df)
pred
```

```python
index = np.where(pred < 0)
index
```

```python
x = df.values
```

```python
plt.scatter(df.iloc[: , 0] , df.iloc[:,1])
plt.scatter(x[index , 0] , x[index , 1] , edgecolors='r')
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles

x,y = make_circles(n_samples = 750 , factor = 0.3 , noise = 0.1)
```

```python
plt.scatter(x[:,0] , x[: , 1])
```

```python
dbscan = DBSCAN(eps = 0.1)
```

```python
dbscan.fit_predict(x)
```

```python
dbscan.labels_
```

```python
plt.scatter(x[:,0] , x[:,1] , c = dbscan.labels_)
```

# Local Outlier Factor

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_moons

x,y = make_moons(n_samples= 300 , noise = 0.1)
```

```python
x_outliers = np.random.uniform(low  = -3 , high =3 , size = (20, 2))
x = np.vstack((x,x_outliers))
```

```python
out = LocalOutlierFactor(n_neighbors =20 , contamination = 0.1)
y_pred = out.fit_predict(x)
```

```python
y_pred
```

```python
#seperate normal data point>>inliers and outlires

x_inliers = x[y_pred == 1]
x_outliers = x[y_pred == -1]

# plotting
plt.scatter(x_inliers[:,0] , x_inliers[:,1] , color ='b' , edgecolor = 'k' , label = 'normal')
plt.title("local outlier  factor on moons datasets")
plt.scatter(x_outliers[:,0] , x_outliers[:,1] , color ='r' , edgecolor = 'k' , label = 'outlier')
plt.legend()
plt.show()
```
